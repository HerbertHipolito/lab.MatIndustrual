```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#Classes e mÃ©todos utilizados#

```python
class doolittle:

  def __init__(self,matrix):

    self.matrix = matrix

  def get_LU(self):

    n=len(self.matrix)
    [l,u]=[np.zeros([n,n]),np.zeros([n,n])]
    
    for j in range(n):
        #print(u[0][j],self.matrix[0][j],l[j][j])
        u[0][j] = self.matrix[0][j]
        l[j][j] = 1
    
    for i in range(1,n):

      l[i][0] = self.matrix[i][0]/u[0][0]
      
    for i in range(1,n):
      
      for j in range(1,n):
        
        if j>=i:
          aux = 0

          for z in range(0,i):

            aux += l[i][z]*u[z][j]
          
          u[i][j] = self.matrix[i][j] - aux

        if j<i:
          aux = 0
        
          for z in range(0,j):

            aux += l[i][z]*u[z][j]
        
          l[i][j] = (self.matrix[i][j]-aux)/u[j][j]

      
    return [l,u]

  def get_y(self,l,b):

    y = np.zeros(len(self.matrix))

    y[0] = b[0]

    for i in range(1,len(b)):

      aux = 0

      for j in range(0,i):
          
        aux += y[j]*l[i][j]

      y[i] = b[i] - aux
      
    return y


  def final_result(self,u,y):

    n = len(y) - 1
    x = np.zeros(n+1)
    #print(y,x)
    x[n] = y[n]/u[n][n]

    for i in range(n-1,-1,-1):
      aux = 0
      for j in range(n,i-1,-1):
        aux += u[i][j]*x[j]
      
      x[i] = (y[i] - aux)/u[i][i]
    
    return x
```

#Projeto 02#

```python
matrix= [[0.866,0,-0.5,0,0,0],[0.5,0,0.866,0,0,0],[-0.866,-1,0,-1,0,0],[-0.5,0,0,0,-1,0],[0,1,0.5,0,0,0],[0,0,-0.866,0,0,-1]]

b = [0,-1000,0,0,0,0]

doolittle_method = doolittle(matrix)

[l,u] = doolittle_method.get_LU()

print(f"Matrix L e U respectivamente\n")

display(pd.DataFrame(l))
display(pd.DataFrame(u))

y=doolittle_method.get_y(l,b)
print(f"\nVetor Y = {y}\n")
x = doolittle_method.final_result(u,y)

print(f"\nresultado final: {x}")
```
